## Adicionando a coluna com os dados relativos

```python
df.columns
```

```python
years = ['2020', '2021', '2022', '2023', '2024']
columns = ['BIRTHS', 'DEATHS']
for year in years:
    pop_column = f'POPESTIMATE{year}'
    for col in columns:
        column = f'{col}{year}'
        new_column = f'{col}PER1000HABITANTS{year}'
        df[new_column] = 0.0 
        for index, row in tqdm(df.iterrows(),total=df.shape[0]):
            value = row[column]
            population = row[pop_column]
            relative_value = (value / population) * 1000  # per 1000 people
            
            df.at[index, new_column] = relative_value
            
df.head()
```

```python
df.to_csv('df_modificado.csv', index=False)
```
